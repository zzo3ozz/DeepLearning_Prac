```python
import tensorflow as tf
print(tf.__version__)
```

```python
X = tf.placeholder(tf.float32, name="X")
Y = tf.placeholder(tf.float32, name="Y")
print(X)
print(Y)
```

```python
x_data = [158.9, 173.6, 174.4, 154.0, 177.7,
          188.3, 183.4, 168.7, 185.1, 156.5,
          160.6, 169.5, 167.9, 154.8, 163.0,
          176.1, 186.7, 172.0, 165.5, 187.5,
          157.3, 158.1, 175.3, 166.3, 153.2]

y_data = [52.015, 66.91, 66.0, 49.275, 69.82,
          79.479, 75.685, 61.569, 76.346, 51.694,
          54.879, 62.276, 61.265, 49.79, 55.808,
          69.12, 78.641, 64.674, 59.633, 78.157,
          50.789, 52.301, 67.076, 59.815, 46.92]

W = tf.Variable(tf.random_normal([1]))
b = tf.Variable(tf.random_normal([1]))

hypothesis = W * X + b
cost = tf.reduce_mean(tf.square(hypothesis - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.00001)
train_op = optimizer.minimize(cost)
```

```python
with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())

  for step in range(1000):
    _, cost_val = sess.run([train_op, cost], feed_dict = {X: x_data, Y: y_data})

    print(step, cost_val, sess.run(W), sess.run(b))
  
  print("\n===test===")
  print("X:183.5 Y:", sess.run(hypothesis, feed_dict={X: 183.5}))
```
